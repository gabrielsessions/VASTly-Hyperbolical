```python
import numpy as np
import pandas as pd
import re
from datetime import date
```

```python
def time_to_float(time):
    if time is None:
        time_val=-86400
    else:
        time_val=0
        time_parts=re.split('[:]',time)
        time_parts.reverse()
        for part in range(0,3):
            time_val=time_val+float(time_parts[part])*60**part
    return time_val
```

```python
def days_since_start(y_m_d):
    days=0
    date_parts=re.split('[-]',y_m_d)

    return((date(int(date_parts[0]),int(date_parts[1]),int(date_parts[2]))-date(2015, 5, 1)).days)
```

```python
raw_data=pd.read_csv('Lekagul Sensor Data.csv')
```

```python
stamp=raw_data['Timestamp'].str.split(expand=True)
```

```python
raw_data['time']=stamp[1]
raw_data['day']=stamp[0]
```

```python
raw_data['vec_time']=raw_data['time'].map(time_to_float)
raw_data['vec_date']=raw_data['day'].map(days_since_start)
```

```python
gate_names=list(raw_data['gate-name'].unique())
```

```python
for gate in gate_names:
    raw_data[gate]=np.nan
for gate in gate_names:
    raw_data.loc[raw_data['gate-name'] == gate, gate] = raw_data.loc[raw_data['gate-name'] == gate, 'vec_time']
```

```python
iten_data_first=raw_data.groupby(['car-id']).first()
iten_data_first.drop(columns=['Timestamp','gate-name'],inplace=True)
iten_data_first.fillna(-86400,inplace=True)
```

```python
iten_data_mean=raw_data.groupby(['car-id']).mean()
iten_data_mean.insert(0,'car-type',iten_data_first['car-type'])
iten_data_mean.fillna(-86400,inplace=True)
```
